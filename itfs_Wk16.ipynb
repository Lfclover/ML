```python
df1=df.drop(['All-cause death count, 2017','All-cause death count, 2018','All-cause death count, 2019','Excess deaths in 2020 per 100,000 population, relative to 2017-2019 average','All-cause death count, 2020'
         ], axis=1)

df1.dropna(inplace=True)
```

```python
fig=px.scatter(df1,x='Population 2020',y='Median age',color='Country name',title='Population of countries with its median age')
fig.show()
```

```python
df2=df1.loc [:,['Country name','COVID-19 deaths per 100,000 population in 2020']]
```

```python
df3=df2.sort_values(ascending=False,by='COVID-19 deaths per 100,000 population in 2020').head(10)
```

```python
df3
```

```python
fig=px.bar(df3,x='Country name',y='COVID-19 deaths per 100,000 population in 2020',title='Top 10 countries with the most deaths per 100,000 population')
fig.show()
```

first place belgium then slovenia etc...

```python
df4=df.loc[:,['Country name','Population 2020','Population 2019']].sort_values(ascending=False,by='Population 2020').head(10)
```

```python
df4
```

```python
fig=px.bar(df4,x='Country name',y=['Population 2020','Population 2019'],barmode='group',title='Top 10 countries with highest population in 2020')

fig.show()
```
