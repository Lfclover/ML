```python
x=df.drop('churn',axis=1)
```

```python
x.head()
```

```python
y=df.churn
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test= train_test_split(x,y,test_size=0.45,random_state=40)
```

```python
from sklearn.preprocessing import StandardScaler

sc=StandardScaler()

X_train= sc.fit_transform(X_train)

X_test= sc.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

Lg=LogisticRegression()

Lg.fit(X_train,y_train)
```

```python
Lg_pred=Lg.predict(X_test)
```

```python
from sklearn import metrics

print(metrics.accuracy_score(y_test,Lg_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier

Rf=RandomForestClassifier(n_estimators=40,random_state=30)

Rf.fit(X_train,y_train)
```

```python
Rf_pred=Rf.predict(X_test)

print(metrics.accuracy_score(y_test,Rf_pred))
```

```python
from sklearn.svm import SVC

clf=SVC(kernel='linear',random_state=30)

clf.fit(X_train,y_train)
```

```python
clf_pred=clf.predict(X_test)


print(metrics.accuracy_score(y_test,clf_pred))
```
