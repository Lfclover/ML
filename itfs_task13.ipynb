```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
top_10=df[["country_id","cm_name"]].groupby("cm_name").agg(['count'])["country_id"]["count"].sort_values(ascending=False)[:10]
```

```python
top_10
```

```python
plt.figure(figsize=(20,15))
sns.barplot(x=top_10.index,y=top_10.values)
plt.title=('Top 10 most reported commodities')
```

```python
top_20=df[["country_id","cm_name"]].groupby("cm_name").agg(['count'])["country_id"]["count"].sort_values(ascending=False)[:20]
top_20
```

```python
plt.figure(figsize=(25,20))
sns.barplot(x=top_20.index,y=top_20.values)
sns.title=('Top 20 commodities')
plt.show()
```

```python
fig=px.bar(df,x='mp_month',y='mp_price',title='The monthly changes in food price')
fig.show()
```

```python
fig=px.bar(df,x='mp_year',y='mp_price',title='The seasonal changes in food price')
fig.show()
```
